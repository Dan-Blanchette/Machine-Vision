# Dense Optical Flow

   - may be mapped to color space visualization

# Algorithms in OpenCV
   - Pyramid Lucas-Kanade
      1. Assume nearby pixels have the same displacement direction
      2. Fixed size window to create a system of equations
      3. Least squares fit to get answer
      4. Image pyramids to deal with large motion
