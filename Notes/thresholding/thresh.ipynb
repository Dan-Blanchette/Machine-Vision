# Truncated Threshold
 - Values higher than threshold -> set to threshold.
 - Otehrwise pixel retains original value.

# To Zero Threshold
 - Set threshold value.
 - If pixel higher than threshold -> keep value.
 - otherwise -> 0.

# To Zero Inverse Threshold
 - Set threshold value
 - If pixel lower than threshold -> keep value.
 - otherwise -> 0.

# Adaptive Thresholding (Adaptive Mean)
 1. Calculates mean or Gaussian-Weighted Sum.
 2. Subtract a consant C.
 3. Apply a form of simple thresholding.

# Ideal Lighting 
1. Diffuse or flat lighting
  - evenly distributed across the environment.
  - shadows can mess with algorithms.

2. flat lighting  is succeptible to glare.
  - good for surface inspection.
  - easier, cheaper to achieve than diffuse lighting.
